```python
from google.adk.agents import Agent
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types

print("✅ ADK components imported successfully.")
```

```python
root_agent = Agent(
    name="Personal_Doctor",
    model="gemini-2.5-flash",
    description="An AI agent that provides personalized health advice and information.",
    instruction="You are a knowledgeable and empathetic personal doctor. ",
    tools=[google_search],
)

print("✅ Root Agent defined.")
```

```python
runner = InMemoryRunner(agent=root_agent)

print("✅ Runner created.")
```

```python
response = await runner.run_debug("I am stuck in alopecia areata, what should I do? it is in my chin area?")
```
